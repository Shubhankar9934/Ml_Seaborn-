```python
globalWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globalWarming_df
```

```python
globalWarming_df = globalWarming_df.drop(columns = ['Country Code','Indicator Name','Indicator Code'], axis = 1).set_index('Country Name')
globalWarming_df.head(6)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df)
```

```python
# setting max and min value in heatmap we use vmin and vmax
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df, vmin = 0 , vmax = 22.5)
```

```python
# Changing the color of heatmap using cmap(color map)
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df, vmin = 0 , vmax = 22.5 , cmap = "coolwarm")
```

```python
# Showing the value of data set in heatmap we use annot param
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df, vmin = 0 , vmax = 22.5 ,
            cmap = "coolwarm", annot = True)
```

```python
annot_arr = (['a00','a01','a02'],['a10','a11','a12'],['a20','a21','a22'],['a31','a32','a33'] )
annot_arr
```

```python
# creating annot value and showing it.
plt.figure(figsize = (16,9))
sns.heatmap(arr_2d, annot = annot_arr , fmt = 's')
```

```python
# Various param of annot keywords agrument
plt.figure(figsize = (16,9))

annot_kws = {"fontsize":10,
             "fontstyle":"italic",
             "color":"k",
             "alpha" : 0.9,
             "rotation":"vertical",
             "verticalalignment":"center",
             "backgroundcolor":"w"}
sns.heatmap(globalWarming_df, vmin = 0 , vmax = 22.5 ,
            cmap = "coolwarm", annot = True , annot_kws = annot_kws,
           linewidth = 4 , linecolor = "k")
```

```python
# if you want to remove the color bar of heatmap you can use cbar param.
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df, cbar = False)
```

```python
# If you want to remove xticks and yticks lable you use these param.
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df, cbar = False , xticklabels = False ,
            yticklabels = False)
```

```python
# color bar heatmap argumnet
cbar_kws = {"orientation":"horizontal",
            "shrink":1,
            "extend":"min",
            'extendfrac':0.1,
            'ticks':np.arange(0,22),
            'drawedges':True}
    
plt.figure(figsize = (16,9))
sns.heatmap(globalWarming_df, cbar_kws = cbar_kws)
```

```python
plt.figure(figsize = (16,9))
ax =sns.heatmap(globalWarming_df)
ax.set(title = "Heatmap of 'who is responsible for global warming'",
      xlabel = "C02 emission (metric tons per capital) Per Year",
      ylabel = "Country Name")
sns.set(font_scale = 5)
```
