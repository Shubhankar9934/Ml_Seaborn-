# How to draw Seaborn Histogram / Seaborn Distplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Loading dataset from github
tips_df = sns.load_dataset("tips")
tips_df
```

```python
# to plotting histogram we use distplot  or sns.distplot()
# Syntax : sns.distplot()
# Signature:
# sns.distplot(
#     a=None,
#     bins=None,
#     hist=True,
#     kde=True,
#     rug=False,
#     fit=None,
#     hist_kws=None,
#     kde_kws=None,
#     rug_kws=None,
#     fit_kws=None,
#     color=None,
#     vertical=False,
#     norm_hist=False,
#     axlabel=None,
#     label=None,
#     ax=None,
#     x=None,
# )
# Docstring:
# DEPRECATED: Flexibly plot a univariate distribution of observations.

# .. warning::
#    This function is deprecated and will be removed in a future version.
#    Please adapt your code to use one of two new functions:

#    - :func:`displot`, a figure-level function with a similar flexibility
#      over the kind of plot to draw
#    - :func:`histplot`, an axes-level function for plotting histograms,
#      including with kernel density smoothing

# This function combines the matplotlib ``hist`` function (with automatic
# calculation of a good default bin size) with the seaborn :func:`kdeplot`
# and :func:`rugplot` functions. It can also fit ``scipy.stats``
# distributions and plot the estimated PDF over the data.

# Parameters
# ----------
# a : Series, 1d-array, or list.
#     Observed data. If this is a Series object with a ``name`` attribute,
#     the name will be used to label the data axis.
# bins : argument for matplotlib hist(), or None, optional
#     Specification of hist bins. If unspecified, as reference rule is used
#     that tries to find a useful default.
# hist : bool, optional
#     Whether to plot a (normed) histogram.
# kde : bool, optional
#     Whether to plot a gaussian kernel density estimate.
# rug : bool, optional
#     Whether to draw a rugplot on the support axis.
# fit : random variable object, optional
#     An object with `fit` method, returning a tuple that can be passed to a
#     `pdf` method a positional arguments following a grid of values to
#     evaluate the pdf on.
# hist_kws : dict, optional
#     Keyword arguments for :meth:`matplotlib.axes.Axes.hist`.
# kde_kws : dict, optional
#     Keyword arguments for :func:`kdeplot`.
# rug_kws : dict, optional
#     Keyword arguments for :func:`rugplot`.
# color : matplotlib color, optional
#     Color to plot everything but the fitted curve in.
# vertical : bool, optional
#     If True, observed values are on y-axis.
# norm_hist : bool, optional
#     If True, the histogram height shows a density rather than a count.
#     This is implied if a KDE or fitted density is plotted.
# axlabel : string, False, or None, optional
#     Name for the support axis label. If None, will try to get it
#     from a.name if False, do not set a label.
# label : string, optional
#     Legend label for the relevant component of the plot.
# ax : matplotlib axis, optional
#     If provided, plot on this axis.

# Returns
# -------
# ax : matplotlib Axes
#     Returns the Axes object with the plot for further tweaking.

# See Also
# --------
# kdeplot : Show a univariate or bivariate distribution with a kernel
#           density estimate.
# rugplot : Draw small vertical lines to show each observation in a
#           distribution.

# Examples
# --------

# Show a default plot with a kernel density estimate and histogram with bin
# size determined automatically with a reference rule:

# .. plot::
#     :context: close-figs

#     >>> import seaborn as sns, numpy as np
#     >>> sns.set_theme(); np.random.seed(0)
#     >>> x = np.random.randn(100)
#     >>> ax = sns.distplot(x)
```

```python
# Plotting Histogram using distplot via sns.distplot() function
sns.distplot(tips_df["size"])
plt.show()
```

```python
sns.distplot(tips_df["total_bill"])
plt.show()
```

```python
sns.distplot(tips_df["tip"])
plt.show()
```

```python
# if you want to change the number of bins in histogram we use bins param
sns.distplot(tips_df["tip"], bins = 100)
plt.show()
```

```python
# If you don't want histogarm only want to show KDE then use hist = False param
sns.distplot(tips_df["tip"], bins = 100  , hist = False)
plt.show()
```

```python
 # if you want to remove or add kde indicator from your histogram the use kde param
sns.distplot(tips_df["tip"],kde = False)
plt.show()
```

```python
sns.distplot(tips_df["tip"],kde = True , rug = True)
```

```python
# to add norm in your histogram we havde to use fit = norm and
# from scipy.stats import norm else you got an error
sns.distplot(tips_df["tip"],kde = True , rug = True, fit = norm)
```

```python
# change the color of histogram we use color param
sns.distplot(tips_df["tip"],kde = True , rug = True, fit = norm, color = "g")
```

```python
# to change the histogarm  in horizontal format we have to use vertical = True param 
sns.distplot(tips_df["tip"],kde = True , rug = True, 
             fit = norm , color = "g", vertical = True)
```

```python
sns.distplot(tips_df["tip"],kde = True , rug = True, 
             fit = norm , color = "g", vertical = True , norm_hist=False)
```

```python
# if you want to change the xlabel then we have to use axlabel to assign the new name
sns.distplot(tips_df["tip"],kde = True , rug = True, 
             fit = norm , color = "g", vertical = False ,
             norm_hist=False ,axlabel = "Total Bill" )
```

```python
sns.distplot(tips_df["tip"],kde = True , rug = True, 
             fit = norm , color = "g", vertical = False ,
             norm_hist=False ,axlabel = "Tip", label = "Tip" )
plt.legend(loc = 4)
plt.title("Histogram to Tip")
```

```python
plt.figure(figsize = (16,9))
sns.set_theme(style='darkgrid')
sns.distplot(tips_df["tip"],kde = True , rug = True, 
             fit = norm , color = "g", vertical = False ,
             norm_hist=False ,axlabel = "Tip", label = "Tip" )
plt.legend()
plt.title("Histogram to Tip")
```

```python
# Shorting Values of dataframe syntax: df.
tips_df.total_bill.sort_values()
```

```python
# Creating Range of bars of histogram using bins
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set_theme(style='darkgrid')
plt.figure(figsize = (16,9))

sns.distplot(tips_df["total_bill"],bins = bins)

plt.xticks()
plt.title("Histogram to Tip")
plt.show()
```

```python
# using param hist_kws and modify the histogram bar accordingly
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set_theme(style='darkgrid')
plt.figure(figsize = (16,9))

sns.distplot(tips_df["total_bill"],bins = bins, hist_kws = {"color":"r",
                                                           'edgecolor':"g"
                                                           ,'linewidth':5
                                                           ,'linestyle':'--'
                                                           ,"alpha":0.9})

plt.xticks()
plt.title("Histogram to Tip")
plt.show()
```

```python
# using param kde_kws and modify the histogram bar accordingly

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set_theme(style='darkgrid')
plt.figure(figsize = (16,9))

sns.distplot(tips_df["total_bill"],bins = bins, hist_kws = {"color":"r",
                                                           'edgecolor':"g"
                                                           ,'linewidth':5
                                                           ,'linestyle':'--'
                                                           ,"alpha":0.9},
            kde_kws = {"color":"m","linewidth":5,"linestyle":"--","alpha" : 0.9})

plt.xticks()
plt.title("Histogram to Tip")
plt.show()
```

```python
# using param rug_kws and modify the histogram bar accordingly

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set_theme(style='darkgrid')
plt.figure(figsize = (16,9))

sns.distplot(tips_df["total_bill"],bins = bins, hist_kws = {"color":"r",
                                                           'edgecolor':"g"
                                                           ,'linewidth':5
                                                           ,'linestyle':'--'
                                                           ,"alpha":0.9},
            kde_kws = {"color":"m","linewidth":5,"linestyle":"--","alpha" : 0.9},
            rug = True,
            rug_kws = {"color":"y","linewidth":3,"linestyle":"--",
                       "alpha" : 0.9})

plt.xticks()
plt.title("Histogram to Tip")
plt.show()
```

```python
# using param fit_kws and modify the histogram bar accordingly

bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set_theme(style='darkgrid')
plt.figure(figsize = (16,9))

sns.distplot(tips_df["total_bill"],bins = bins, hist_kws = {"color":"r",
                                                           'edgecolor':"g"
                                                           ,'linewidth':5
                                                           ,'linestyle':'--'
                                                           ,"alpha":0.9},
             kde= False,
             fit = norm,
            fit_kws = {"color":"m","linewidth":12,"linestyle":"--","alpha" : 0.9},
            rug = True,
            rug_kws = {"color":"y","linewidth":3,"linestyle":"--",
                       "alpha" : 0.9, 'label' : "TB"})

plt.xlabel("total bill", fontsize = 15)
plt.ylabel("tips",fontsize = 15)
plt.xticks()
plt.title("Histogram to Tip", fontsize = 28)
plt.show()

```

```python
# Plotting Multiple Histogram in same bar
sns.distplot(tips_df["total_bill"],bins = 9)
sns.distplot(tips_df["tip"],bins = 9)
sns.distplot(tips_df["size"],bins = 9)
```
